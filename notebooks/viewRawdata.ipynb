```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_excel(r'..\data\rawData\train_nor_811.xlsx')
```

```python
df_train['Emotion'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='Emotion', data=df_train, palette='Set2')
plt.title('Biểu đồ phân phối các nhãn cảm xúc trong tập dữ liệu huấn luyện')
plt.xlabel('Emotion')
plt.ylabel('Số lượng')
plt.xticks(rotation=45)

# Lưu biểu đồ trước khi show
plt.savefig(r'E:\EmotionAnalysis\Emotions-Analysis\outputs\figures\Static_emotion.png', dpi=300, bbox_inches='tight')

plt.show()
```
